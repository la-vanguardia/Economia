```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import animation
from IPython.display import HTML
import seaborn as sns
```

```python
def graficar_isocuanta( Ls, Ks, colors , legend ):
    fig = plt.figure( figsize=(16,9) )
    ax = fig.gca()
    ax.grid()
    ax.legend( legend )
    for L, K, color in zip( Ls, Ks, colors ):
        ax.plot( L, K, color=color, marker='.', ms=20, lw=3)
        
    return fig
```

```python
plt.rcParams.update( {'font.size': 22} )
```

```python
K = [ 10, 6, 4.5, 3.5, 3, 2.7 ]
L = [ i+2 for i in range( 6 )]
```

```python
fig = graficar_isocuanta( L, K, 'red', 0 )
```
