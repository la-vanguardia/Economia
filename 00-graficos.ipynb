```python
def TMST( L, K ):
    TMST = [ np.nan ]
    for i in range( len(L)-1 ):
        delta_L = L[i] - L[i-1]
        delta_K = K[i] - K[i-1]
        TMST.append( -delta_K / delta_L )
    return TMST
```

```python
TMST0 = TMST( Ls[0], Ks[0] )
TMST1 = TMST( Ls[1], Ks[1] )
```

```python
f = open( './datos/punto4.txt', 'w' )
for tmst in TMST0:
    f.write( '\t' )
    f.write( str( tmst ) )
f.write( '\n' )
for tmst in TMST1:
    f.write( '\t' )
    f.write( str( tmst ) )
f.close()
```
